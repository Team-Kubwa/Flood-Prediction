```python
import matplotlib.pyplot as plt

# Plotting a histogram for 'TopographyDrainage'
plt.figure(figsize=(10, 6))
plt.hist(train['TopographyDrainage'], bins=range(int(train['TopographyDrainage'].min()), int(train['TopographyDrainage'].max()) + 2), edgecolor='black')
plt.title('Distribution of Topography Drainage')
plt.xlabel('Topography Drainage')
plt.ylabel('Frequency')
plt.xticks(range(int(train['TopographyDrainage'].min()), int(train['TopographyDrainage'].max()) + 1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```
